## Multicolinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Year','Mileage','EngineV']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif['Features'] = variables.columns
```

```python
vif
```

## vif = 1 no colinearity 1<vif<5 perfect colinearity vif>10 unacceptable can be ignored

```python
data_no_multicolinearity = data_cleaned.drop(['Year'], axis = 1)
```

### create dummies

```python
data_with_dummies = pd.get_dummies(data_no_multicolinearity, drop_first = True)
data_with_dummies.head()
```

```python
data_with_dummies.columns.values
```

```python
cols = ['Log Price','Mileage', 'EngineV',  'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

### Linear Regression

```python
targets = data_preprocessed['Log Price']
inputs = data_preprocessed.drop(['Log Price'], axis = 1)
```

### scale the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

### train test split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size= 0.2, random_state = 365)
```

### Create Regression

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat)
plt.xlabel('Targets (y_train)' , size = 10)
plt.ylabel('predictions (y_hat)' , size = 10)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
sns.distplot(y_train-y_hat)
plt.title('Residual PDF', size = 10)
```

```python
reg.score(x_train,y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns = ['features'])
reg_summary['weights'] = reg.coef_
reg_summary
```

### Testing

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test, y_hat_test, alpha = 0.2)
plt.xlabel('Targets (y_test)' , size = 10)
plt.ylabel('predictions (y_hat_test)' , size = 10)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test), columns = ['prediction'])
df_pf.head()
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf.head()
```

```python
y_test = y_test.reset_index(drop = True)
y_test.head()
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf.head()
```

```python
df_pf['residual'] = df_pf['Target'] - df_pf['prediction']
df_pf['diff%'] = np.abs(df_pf['residual']/df_pf['Target']*100)
```
